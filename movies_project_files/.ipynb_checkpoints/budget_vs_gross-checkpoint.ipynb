# Box Office Performance and Budget Analysis

## Is there a relationship between a movie's budget, box office performance, and ratings?

### Budget vs. Box Office Earnings:
- An r-value of 0.5476 indicates a moderate positive linear relationship between the budget and gross of movies. As the budget increases, there tends to be a moderate increase in gross, but it's not as strong as a higher correlation would suggest.
- The moderate correlation suggests that while there is a tendency for movies with higher budgets to have higher gross earnings, there is significant variability. In other words, there are instances where movies with lower budgets might still perform well in terms of gross earnings, and vice versa.

### Budget vs. Ratings:
- An r-value of 0.0051 indicates an almost non-existent linear relationship between the budget allocated for a movie and its rating. Changes in the budget don't show a significant pattern or impact on the movie ratings.
- It suggests that the budget spent on a movie does not significantly influence or predict the rating it receives. Therefore, when aiming for a higher movie rating, other factors beyond budget allocation need to be considered and prioritized; like quality of the script, acting performances, audience preferences and so on.

### Relationship between Average Budget and Average Gross of Movies over the Years:
- A Pearson correlation coefficient of approximately 0.94 suggests a strong positive linear relationship between the average budget and average gross of movies over the years. The p-value associated with this correlation is extremely small (3.26e-19), indicating strong evidence against the null hypothesis that there is no correlation between the variables. This very low p-value suggests that the correlation coefficient is statistically significant.
- This strong positive correlation over the years could reflect changes in audience preferences, production quality or other factors that have led to increased box office revenues for movies with higher budgets. While a high budget might correlate with higher gross, it also signifies a higher financial risk if the movie doesn't perform well.
- Although there's a strong correlation, it's important to note that correlation doesnâ€™t imply causation. Other factors, such as marketing strategies, script quality, casting, release timing, etc., also play crucial roles in a movie's success at the box office.


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import linregress
```

```python
# Load cleaned movie data CSV file
movie_data_df = pd.read_csv("../Resources/movies_cleaned.csv")

# Display sample data
movie_data_df.head()
```

```python
# Pull relavent data for Budget vs gross earnings into separate dataframe
rel_movie_data_df = movie_data_df[['Movie_Name', 'Year', 'Rating_Score', 'Votes', 'Budget', 'Gross']]

# Display new DataFrame
rel_movie_data_df.head()
```

```python
rel_movie_data_df.count()
```

```python
# Building scatter plot for budget vs. gross and Linear `regression Plot

x_values = rel_movie_data_df["Budget"]
y_values = rel_movie_data_df["Gross"]

slope, intercept, rvalue, pvalue, stderr = linregress(x_values, y_values)

regress_values = x_values * slope + intercept

plt.figure(figsize=(10,6))

plt.scatter(x_values, y_values, c="teal", alpha=0.5)

plt.plot(x_values, regress_values, color="red")

plt.xlabel("Budget ($ 100 Million)")
plt.ylabel("Gross Earnings ($ Billion)")
plt.title("Budget vs. Gross Earnings")

line_eq = f"y = {round(slope, 2)}x + {round(intercept, 2)}"
plt.annotate(line_eq, (0,2500000000), fontsize=12, color="red")

correlation = f"Correlation: {round(rvalue, 2)}"
print(f"The r-value is: {rvalue**2}")

plt.savefig("Images/Fig5.1.png")

plt.show()
```

```python
# Building scatter plot for budget vs. rating_score `regression Plot

x_values = rel_movie_data_df["Budget"]
y_values = rel_movie_data_df["Rating_Score"]

slope, intercept, rvalue, pvalue, stderr = linregress(x_values, y_values)

regress_values = x_values * slope + intercept

plt.figure(figsize=(10,6))

plt.scatter(x_values, y_values, c="orange", alpha=0.5)

plt.plot(x_values, regress_values, color="red")

plt.xlabel("Budget ($ 100 Million)")
plt.ylabel("Rating Score")
plt.title("Budget vs. Rating Score")

line_eq = f"y = {round(slope, 2)}x + {round(intercept, 2)}"
plt.annotate(line_eq, (200000000,3), fontsize=12, color="red")

correlation = f"Correlation: {round(rvalue, 2)}"
print(f"The r-value is: {rvalue**2}")

plt.savefig("Images/Fig5.2.png")

plt.show()
```

```python
average_by_year = rel_movie_data_df.groupby('Year').agg({'Budget': 'mean', 'Gross': 'mean'})

# Displaying summary statistics in another DataFrame
summary_statistics = average_by_year.describe()

print("Average Budget and Gross by Year:")
print(average_by_year)
print("\nSummary Statistics:")
print(summary_statistics)
```

```python
average_by_year = average_by_year.rename(columns={'Budget': 'average_budget', 'Gross': 'average_gross'})
average_by_year.reset_index()
```

```python
# Plotting Line Chart for Average Budget and Gross Income of movies over the years
yearly_averages_df = pd.DataFrame(average_by_year)

plt.figure(figsize=(10, 6))
plt.plot(yearly_averages_df['average_budget'], label='Average Budget', color='salmon')
plt.plot(yearly_averages_df['average_gross'], label='Average Gross', color='teal')
plt.xlabel('Year')
plt.ylabel('Amount ($100 million)')
plt.title('Average Budget and Average Gross by Year')
plt.legend()
plt.grid(True)
plt.savefig("Images/Fig5.3.png")

plt.show()
```

```python
from scipy.stats import pearsonr

correlation_coefficient, p_value = pearsonr(yearly_averages_df['average_budget'], yearly_averages_df['average_gross'])

print(f"Pearson correlation coefficient: {correlation_coefficient}")
print(f"P-value: {p_value}")

# Interpretation based on p-value
alpha = 0.05  # significance level
if p_value < alpha:
    print("There is a significant relationship between average budget and average gross of movies over the years.")
else:
    print("There is no significant relationship between average budget and average gross of movies over the years.")
```

```python
average_by_country = movie_data_df.groupby('Country').agg({'Budget': 'mean', 'Gross': 'mean'})

# Displaying summary statistics in another DataFrame
summary_statistics1 = average_by_country.describe()

print("Average Budget and Gross by Country:")
print(average_by_country)
print("\nSummary Statistics:")
print(summary_statistics1)
```

```python
country_averages_df = pd.DataFrame(average_by_country).reset_index()
country_averages_df
```

```python
import numpy as np

# Extracting data
countries = country_averages_df['Country']
budget = country_averages_df['Budget']
gross = country_averages_df['Gross']

# Setting up the positions for the bars
bar_width = 0.35
index = np.arange(len(countries))

# Creating the grouped bar chart
plt.figure(figsize=(12, 8))

plt.bar(index, budget, bar_width, label='Budget')
plt.bar(index + bar_width, gross, bar_width, label='Gross')

plt.xlabel('Country')
plt.ylabel('Amount')
plt.title('Budget and Gross by Country')
plt.xticks(index + bar_width / 2, countries, rotation=90)
plt.legend()
plt.tight_layout()

plt.show()
```
