```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import linregress
```

```python
# Load cleaned movie data CSV file
movie_data_df = pd.read_csv("../Resources/movies_cleaned.csv")

# Display sample data
movie_data_df.head()
```

```python
# Pull relavent data for Budget vs gross earnings into separate dataframe
rel_movie_data_df = movie_data_df[['Movie_Name', 'Year', 'Rating_Score', 'Votes', 'Budget', 'Gross']]

# Display new DataFrame
rel_movie_data_df.head()
```

```python
rel_movie_data_df.count()
```

```python
# Building scatter plot for budget vs. gross and Linear `regression Plot
# Northern Hemisphere
x_values = rel_movie_data_df["Budget"]
y_values = rel_movie_data_df["Gross"]

slope, intercept, rvalue, pvalue, stderr = linregress(x_values, y_values)

regress_values = x_values * slope + intercept

plt.figure(figsize=(10,6))

plt.scatter(x_values, y_values, c="teal", alpha=0.5)

plt.plot(x_values, regress_values, color="red")

plt.xlabel("Budget ($ 100 Million)")
plt.ylabel("Gross Earnings ($ Billion)")
plt.title("Budget vs. Gross Earnings")

line_eq = f"y = {round(slope, 2)}x + {round(intercept, 2)}"
plt.annotate(line_eq, (0,2500000000), fontsize=12, color="red")

correlation = f"Correlation: {round(rvalue, 2)}"
print(f"The r-value is: {rvalue**2}")

plt.show()
```

```python
# Building scatter plot for budget vs. rating_score `regression Plot
# Northern Hemisphere
x_values = rel_movie_data_df["Budget"]
y_values = rel_movie_data_df["Rating_Score"]

slope, intercept, rvalue, pvalue, stderr = linregress(x_values, y_values)

regress_values = x_values * slope + intercept

plt.figure(figsize=(10,6))

plt.scatter(x_values, y_values, c="orange", alpha=0.5)

plt.plot(x_values, regress_values, color="red")

plt.xlabel("Budget ($ 100 Million)")
plt.ylabel("Rating Score")
plt.title("Budget vs. Rating Score")

line_eq = f"y = {round(slope, 2)}x + {round(intercept, 2)}"
plt.annotate(line_eq, (200000000,3), fontsize=12, color="red")

correlation = f"Correlation: {round(rvalue, 2)}"
print(f"The r-value is: {rvalue**2}")

plt.show()
```
