# OMDB Movie Ratings and Trends
---
### Hypothesis Testing

#### Null Hypothesis
    - The time series is non-stationary, so ratings vary over time through trends and seasonality.
    
#### Alternative Hypothesis
    - The times series is stationary, so ratings do not vary over time and fluctuate around a constant level.


### Analysis

Explore the rating scores between IMDB and Metascore over the years:
- The ratings given by Metascore and IMDB are provided in two different ways. The Metascore is a weighted average of reviews from critics and publications for a given movie along with user scores. IMDB, meanwhile, is open to anyone with an IMDB account. IMDB movie scores can vary wildly depending on the user, though the website does try and ensure extreme scores don't adversely affect the rating. Since it doesn't rely on critical evaluation, the score it's far more of an indicator of popularity than it is actual quality.
- Starting with the IMDb line, we can see that most ratings are between 6 and 8.5. The highest point of the lines shows in the earlier years, with the line gradually settling between 6.5 and 7 from 1978 onwards before suddenly spiking back up from 2020 and onwards.
- The Metascore line is distributed more widely between 5.5 and 10. It follows a similar pattern to the IMDB line, though it rises higher than the IMDB line in the earlier years. It then falls below the IMDB line from 1978 onwards, before spiking back up in 2020. 
- The p-value for both Metascore and IMDB is greater than or equal to 0.05, so the null hypothesis would not be rejected and the alternative hypothesis would be rejected to show that ratings do vary through trends, and it would be affected by other factors.

Explore the number of votes given to movies on IMDB over the years:
- When looking along the line, there are noticeable peaks throughout certain years where there are a large number of voters on average. With IMDB having started as a company in 1990, it may be that these older movies have had more time to accumulate ratings. It is also the case that classic films continue to be highly regarded by people, with classics like Spartacus/Psycho in 1960, Grease in 1978 and The Lion King in 1994. While taking these peaks into account, it can also be seen that the line is generally tending positively, getting slowly higher over the years before a huge spike in 2020 which can also be seen with the rating scores.

Explore the distribution of the highest grossing movie genres over the years:
- Near to the start of the figure, the lines for the genres were at relatively similar levels showing a wide range in high grossing films. This then quickly spreads out, with Action, Comedy, Drama and Adventure increasing dramatically which shows that these genres have become a big draw for audiences and also highlights the growing number of movies that are coming out year upon year.

```python
 # Import Dependencies
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
%matplotlib inline
```

```python
# Create a reference to the CSV and import it into a Pandas DataFrame
OMDBmoviedata_df = pd.read_csv('../Resources/OMDBMovieData.csv', thousands=',')
OMDBmoviedata_df.head()
```

```python
#Drop the columns that are not required
OMDBmoviedata_df = OMDBmoviedata_df.drop(columns=['Runtime', 'Writer', 'Actors', 'Plot', 'Language', 'Country', 'Awards', 'Poster', 'Type', 'DVD', 'Production', 'Website', 'Response', 'totalSeasons', 'Error'])
OMDBmoviedata_df.info()
```

```python
#Checking the number of null values
OMDBmoviedata_df.isnull().sum()
```

```python
#Drop the rows with N/A
columns_to_check = ['Title', 'Year', 'Rated', 'Released', 'Genre', 'Director', 'Ratings', 'Metascore', 'imdbRating', 'imdbVotes', 'imdbID', 'BoxOffice']
OMDBmoviedata_df.dropna(subset=columns_to_check, inplace=True)
OMDBmoviedata_df
```

```python
# Checking the number of null values again
OMDBmoviedata_df.isna().sum()
```

```python
#Create Summary Statistics table
OMDBmoviedata_df.describe()
```

```python
#Gather the average imdbRating per year
imdb_year_mean = OMDBmoviedata_df.groupby('Year')['imdbRating'].mean().reset_index()
imdb_year_mean = imdb_year_mean.sort_values(by='Year')
print(imdb_year_mean)
```

```python
#Gather the average Metascore per year
Metascore_year_mean = OMDBmoviedata_df.groupby('Year')['Metascore'].mean().reset_index()
Metascore_year_mean = Metascore_year_mean.sort_values(by='Year')
Metascore_year_mean['Metascore'] = Metascore_year_mean['Metascore'].div(10)
print(Metascore_year_mean)
```

```python
#Plot graph to compare IMDB and Metascore over the years, and use Augmented Dickeyâ€“Fuller test to test hypothesis
plt.figure(figsize = (10,6))
plt.plot(imdb_year_mean['Year'], imdb_year_mean['imdbRating'], label = "IMDB")
plt.plot(Metascore_year_mean['Year'], Metascore_year_mean['Metascore'], label = "Metascore")
x = imdb_year_mean['imdbRating'].values
result = adfuller(x)
y = Metascore_year_mean['Metascore'].values
result2 = adfuller(y)
print('IMDB -')
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %3.f' % (key, value))
print('Metascore -')
print('ADF Statistic: %f' % result2[0])
print('p-value: %f' % result2[1])
print('Critical Values:')
for key, value in result2[4].items():
    print('\t%s: %3.f' % (key, value))
plt.title('Ratings over the Years: Comparison of IMDB and Metascore')
plt.xlabel('Year')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.legend()
plt.savefig("../Images/Fig1.1.png")
plt.show()
```

```python
#Gather the average imdb vote per year
imdb_vote_mean = OMDBmoviedata_df.groupby('Year')['imdbVotes'].mean().reset_index()
imdb_vote_mean = imdb_vote_mean.sort_values(by='Year')
print(imdb_vote_mean)
```

```python
#Plot graph to show IMDB votes over the years
plt.figure(figsize=(10, 6))
plt.plot(imdb_vote_mean['Year'], imdb_vote_mean['imdbVotes'])
plt.title('imdbVoting over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Votes')
plt.xticks(rotation=90)
plt.grid(visible=True)
plt.savefig("../Images/Fig1.2.png")
plt.show()
```

```python
#Split the genre column and put each genre in sperate column
OMDBmoviedata_df[["Genre_1", "Genre_2", "Genre_3"]] = OMDBmoviedata_df["Genre"].str.split(', ', expand=True)
OMDBmoviedata_df
```

```python
#Get the number of each genre per year from the first genre column
genre_count1 = OMDBmoviedata_df.groupby('Year')['Genre_1'].value_counts().reset_index()
genre_count1
```

```python
#Get the number of each genre per year from the second genre column
genre_count2 = OMDBmoviedata_df.groupby('Year')['Genre_2'].value_counts().reset_index()
genre_count2
```

```python
#Get the number of each genre per year from the third genre column
genre_count3 = OMDBmoviedata_df.groupby('Year')['Genre_3'].value_counts().reset_index()
genre_count3
```

```python
#Reshape the dataframe for Genre 1
genre_count1 = genre_count1.pivot(index='Year', columns='Genre_1', values='count')
genre_count1 = genre_count1.rename_axis(None, axis=1)
genre_count1
```

```python
#Reshape the dataframe for Genre 2
genre_count2 = genre_count2.pivot(index='Year', columns='Genre_2', values='count')
genre_count2 = genre_count2.rename_axis(None, axis=1)
genre_count2
```

```python
#Reshape the dataframe for Genre 3
genre_count3 = genre_count3.pivot(index='Year', columns='Genre_3', values='count')
genre_count3 = genre_count3.rename_axis(None, axis=1)
genre_count3
```

```python
#Concatenate all the genre dataframes into one dataframe, and sum all the values under their respective columns
TotalGenre = pd.concat([genre_count1, genre_count2, genre_count3])
TotalGenre = TotalGenre.groupby(level=0).sum()
TotalGenre
```

```python
#Plot graph to compare number of movies per genre over the years
ax = TotalGenre.plot(y = TotalGenre.columns, ylabel = 'Number of Movies', use_index=True, figsize=(15, 10), title = 'Comparison of Genres over the Years', colormap="tab20")
ax.figure.savefig("../Images/Fig1.3.png")
ax.set(xlabel=None)
plt.show()
```
