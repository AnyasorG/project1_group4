```python
import requests
import json
from config import API_Key
from config import API_Key2
from config import API_Key3
from pprint import pprint

import pandas as pd
from pathlib import Path
```

```python
movies_cleaned_csv = Path("movies_cleaned.csv")
```

```python
movies_cleaned_df = pd.read_csv(movies_cleaned_csv)
movies_cleaned_df.head()
```

```python
sorted_movies_cleaned_df = movies_cleaned_df.sort_values(by='Gross', ascending=False)
sorted_movies_cleaned_df.head()
```

```python
movieTitles = list(sorted_movies_cleaned_df.Movie_Name)
movieYears = [sorted_movies_cleaned_df.Release_date.iloc[x].split("-")[0] for x in range(0, len(movieTitles))]
FullMovie = list(zip(movieTitles, movieYears))
FullMovie
```

```python
TopMovieTitles = FullMovie[:3000]
TopMovieTitles[2999]
```

```python
size = 1000
TopMovie1, TopMovie2, TopMovie3 = [TopMovieTitles[i:i+size] for i in range(0, len(TopMovieTitles), size)]
print(TopMovie1[999])
print(TopMovie2[0])
print(TopMovie3[999])
```

```python
url = "http://www.omdbapi.com/?t="
API_Key = "&apikey=" + API_Key
API_Key2 = "&apikey=" + API_Key2
API_Key3 = "&apikey=" + API_Key3
```

```python
responses = []
for mov in TopMovie1:
    movieInfo = requests.get(f'{url}{mov[0]}&year={mov[1]}{API_Key}').json()
    responses.append(movieInfo)
    
responses
```

```python
for mov in TopMovie2:
    movieInfo2 = requests.get(f'{url}{mov[0]}&year={mov[1]}{API_Key2}').json()
    responses.append(movieInfo2)

len(responses)
```

```python
for mov in TopMovie3:
    movieInfo3 = requests.get(f'{url}{mov[0]}&year={mov[1]}{API_Key3}').json()
    responses.append(movieInfo3)

len(responses)
```

```python
moviedata_pd = pd.DataFrame(responses)
```

```python
filepath = Path('OMDBmoviedata.csv')
moviedata_pd.to_csv(filepath, index=False)
```
